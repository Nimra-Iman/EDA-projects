```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
dataset=pd.read_csv("Diwali_Sales_Data.csv", encoding='latin1')
dataset.drop(columns="Status", inplace=True)
dataset.drop(columns="unnamed1", inplace=True)
print(dataset.isnull().sum())               
```

```python
dataset["Amount"]=dataset["Amount"].fillna(dataset["Amount"].mean())
dataset.head()
```

```python
print(dataset.isnull().sum())
```

```python
bars=sns.countplot(x="Gender", data=dataset)
for i in bars.containers:
    bars.bar_label(i)
    
plt.show()     # now females are more than males then we can attract more females via social media platform such as instagram 
# where there is female communities
```

```python
sns.countplot(x="Age Group",hue="Gender", data=dataset)
plt.show()
```

```python
sns.countplot(x="Marital_Status", data=dataset)
plt.show()
```

```python
plt.figure(figsize=(21,7))
max_order=dataset.groupby("State")["Orders"].sum().reset_index().sort_values(by="Orders", ascending=False)
sns.barplot(x="State",y="Orders" ,data=max_order)
plt.show()
```

```python
plt.figure(figsize=(23,7))
sns.countplot(x="Product_Category", data=dataset)
plt.show()
```

```python
# Both married and unmaried people shop more Clothing and Apparel as compared to other product catagory
plt.figure(figsize=(21,7))
sns.countplot(x="Product_Category" , data=dataset[dataset["Marital_Status"]==1])
plt.show()
```

```python
plt.figure(figsize=(21,8))
# sns.barplot(x="Product_Category", y="Orders", data=dataset)  #agar esy graph show kryn to bht s bars ki height y-axis k point vali
# values pr stop krti yani Auto ki peak 2.5 pr stop krti yani orders 2.5 ay, to y 2.5 asal m average h yani in average auto k 2.5 orders ay ab may be
# auto k kuch orders 4 th, kuch 2 th , kuch 3 th kuch 1 th to un sab ka average 2.5 a gya.

max_order=dataset.groupby("Product_Category")['Orders'].sum().reset_index().sort_values(by="Orders", ascending=False)
sns.barplot(x="Product_Category", y="Orders", data=max_order)  #ab is s sum of orders pta lgy ga, yani ab bta rha h k sab s ziada orders clothing
# and apparl k ayn hn

plt.show()    
```

```python
max_amount=dataset.groupby("Product_Category")['Amount'].sum().reset_index().sort_values(by="Amount", ascending=False)
plt.figure(figsize=(20,7))
sns.barplot(x="Product_Category", y="Amount", data=max_amount)  #ab is s sum of amount pta lgy ga, yani ab bta rha h k sab s ziada pesa kis catagory
# s aya h

plt.show()  
```

```python
sns.countplot(x="Gender", hue="Marital_Status", data=dataset)
plt.show()   #both married and unmerried females shop more as compared to marriend and unmaried males respectively.
```

```python
plt.figure(figsize=(15,5))
max_amount=dataset.groupby("Occupation")["Amount"].sum().reset_index().sort_values(by="Amount", ascending=False)
sns.barplot(x="Occupation", y="Amount", data=max_amount)
```

```python
Females, especially aged 26-35, shop more than males.
Unmarried individuals shop more than married ones.
Clothing and Apparel have the highest orders, but Food generates the most revenue.
Uttar Pradesh leads in order volume.
IT sector employees spend the most.
```
