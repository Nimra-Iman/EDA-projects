```python
Estimation: getting some value (like mean, proportion, regression coefficients) from a poplulation
using some sample from it 

Inference: making conclusion or decision( like mean proportion, regression coefficients) from population 
usiong some sample from it and showing confidence to the value. 
    --> 'Inference = Make a conclusion + measure how sure you are'


One-line difference
Estimation: "Whatâ€™s the value?"
Inference: "Is this value significant? How sure are we?"
```

```python
( NOT SURE ABOUT THE CORRECTNES OF CONCEPT, difference exist)
Parameter and underlying parameter: 
A parametr is the characteristic of a poplulation that we get from a sample e.g: averaghe height of a 
population. we cannot get the exact value, we only estimate it from a sample.

Underlying parameters: these are the true parameters that describe the excat value of a poplulation. e.g: to 
calculate average height of a polulation, we calculate the height of each person and then calculate mean from it.
This is impossible to get. So underlying parameters are hidden values that are impossible to get and we 
can infer these values from a parametr with some confidence value.
```

```python
Goal of statistical inference: is about understanding the data generating process.

There are 2 models of statistical inference: 
1- parametric
2- non-parametric 

--> Parametric: In parametric we make some assumptions about the data and calulate 
some parameters from data such as slope of data, or
mean etc of data, SUCH AS determing  y = mx+c by assuming that data follows normal distribution. They have finite number of 
parametrs and we need to define the coefficients according to features. In this case, the 'y' value depend on parametr m and c.

--> Non-parametric: we don't make assumptions and we determine about data such as determining the histogram of the data
to determine the distributuion of this data. We don';t have parametrs in this case.
Non-parametric does not mean we don't know anything about data.
```

```python
CORRELATION VS CAUSATION
Corelation: things happen in a related way. One variable is directly related to the other.
e.g: ice cream sales and swimming rate both go up in summer. These 2 variables ice cream sales and 
swimming rate are corelated.

Causation: one variable become the cause of other variable. 
e.g: Pushing a glass off a table causes it to fall and break.
```

```python
spurious corelation and confounding variable
In both these terms, there is a hidden factor that create relationshio between variables but difference between 
them is following:

SPURIOUS CORRELATION: There is no real relationship between variables. It is just hidden factor that tries to 
relate them.
e.g: 'Ice cream sales and shark attacks both rise in summer'. Summer is a hidden factor that creates relationship
between these variables. although there is no real relationship between ice cream and shark attack.

Confounding Variable: the hidden factor affect both the cause and the affect and make the relation look
different. Although there is a direct relationship between variables but hidden factors confuses the results.
e.g: 'does doing exercise reduces weight'. This relationship is real but hidden factor is diet. persons who do exercise tend to 
eat healthy, and healthy eating itself reduces weight. So, this hidden factor diet confuses the results. if we do not
consider this hidden factor, we might think weight is only reduced by doing execrcise. 

```
